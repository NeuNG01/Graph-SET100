```python
import pandas as pd
import sys
import starfishX as sx
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'figure.max_open_warning': 0})
print(f'pandas version: {pd.__version__}')
from datetime import datetime
str (datetime.now())
SET = 'https://www.settrade.com/C13_MarketSummary.jsp?order=Y'
MAI = 'https://www.settrade.com/C13_MarketSummary.jsp?order=Y&market=mai'

dfs = pd.read_html(SET,
                   match='หลักทรัพย์',
                   header=[0])#&market=mai
len(dfs)
df = dfs[1]
symple = df['หลักทรัพย์'].unique()
symple 
```

```python
for i in symple:
    try:
        df = sx.loadHistData_v2(i,start='2021-01-01',end='2022-12-31')
        df[['Close']].plot(grid=True).set_title(i)
    except:
        print(sys.exc_info())
```
